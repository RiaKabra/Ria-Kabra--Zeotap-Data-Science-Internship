#### Detailed Analysis of Product Dataset for Business Insights

Splitting the ProductName into two parts - Company Name, and Name of the product

```python
def split_first(col):
    col = str(col)  
    return col.split()[0] 

def split_second(col):
    col = str(col)  
    return col.split()[1] 

df_prod['CompanyName'] = df_prod['ProductName'].apply(split_first)
df_prod['ProductDetail'] = df_prod['ProductName'].apply(split_second)
```

```python
company_counts = df_prod['CompanyName'].value_counts()
plt.figure(figsize=(10, 5))
company_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Products by Company')
plt.xlabel('Company Name')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()

product_counts = df_prod['ProductDetail'].value_counts()
plt.figure(figsize=(10, 5))
product_counts.plot(kind='bar', color='orange')
plt.title('Most Sold Product Details')
plt.xlabel('Product Detail')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()

categories = df_prod['Category'].nunique()
print(f"Total number of unique categories: {categories}")


type_prod_counts = df_prod['Category'].value_counts()
plt.figure(figsize=(10,5))
type_prod_counts.plot(kind = 'bar',color = "Pink")
plt.title('Most Sold Products by Category')
plt.show()
```

#### Detailed Analysis of Transaction Dataset for Business Insights

Joining Customers and Products datasets

```python
df_tran = df_tran.merge(df_cust, on='CustomerID', how='left')
df_tran = df_tran.merge(df_prod, on='ProductID', how='left')
```

```python
region_revenue = df_tran.groupby('Region')['TotalValue'].sum()
plt.figure(figsize=(8, 6))
region_revenue.plot(kind='bar', color='green')
plt.title('Revenue Contribution by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue (USD)')
plt.show()

top_products = df_tran.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
top_products.plot(kind='bar', color='purple')
plt.title('Top Products by Sales Volume')
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.show()
```

#### Detailed Analysis of Customers Dataset for Business Insights

Customers by region and Signup over the years

```python
region_counts = df_cust['Region'].value_counts()

plt.figure(figsize=(8, 6))
region_counts.plot(
    kind='pie',
    startangle=90,
    colors=['gold', 'lightblue', 'pink', 'lightgreen', 'orange'],  
    autopct='%1.1f%%'
    
)
plt.title('Customer Distribution by Region')
plt.ylabel('')  
plt.show()

df_cust['SignupYear'] = df_cust['SignupDate'].dt.year
signup_trends = df_cust['SignupYear'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
signup_trends.plot(kind='line', marker='o', color='red')
plt.title('Customer Signup Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Signups')
plt.grid()
plt.show()
```
